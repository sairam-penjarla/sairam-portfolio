```python
import os

root = "courses"

def to_snake_case(name: str) -> str:
    """Convert a string to snake_case"""
    name = name.replace(" ", "_").replace("-", "_").lower()
    return name

# Walk through all categories and courses
categories = os.listdir(root)
for category in categories:
    category_path = os.path.join(root, category)
    if not os.path.isdir(category_path):
        continue

    courses = os.listdir(category_path)
    for course in courses:
        course_path = os.path.join(category_path, course)
        if not os.path.isdir(course_path):
            continue

        files = os.listdir(course_path)
        for file in files:
            if not file.endswith(".md"):
                continue

            # File paths and metadata
            file_path = os.path.join(course_path, file)
            title = file[:-3].lower()  # remove .md and lower
            py_file_name = file.replace(".md", ".py")
            py_file_path = os.path.join(course_path, py_file_name)
            course_snake = to_snake_case(course)

            # Generate Python file content
            content = f'''import os

def get_markdown_content():
    """Reads and returns the latest content of the markdown file"""
    md_path = os.path.join(os.path.dirname(__file__), "{file}")
    with open(md_path, "r", encoding="utf-8") as f:
        return f.read()

course_data = {{
    "title": "{title}",
    "file_name": "{file}",
    "category": "{category}",
    "course": "{course_snake}",
    "imageUrl": "",
    "date": "Duration: 4 Weeks",
    "body": get_markdown_content()
}}
'''

            # Write the .py file
            with open(py_file_path, "w", encoding="utf-8") as f:
                f.write(content)

            print(f"✅ Created: {py_file_path}")
```

```python
import os

# Base folder containing all cheatsheet folders
BASE_DIR = "cheatsheets"

# Constants
IMAGE_URL = "https://iamsairamsa.blob.core.windows.net/static/images/thumbnails/img_7.jpg"
DURATION = "Duration: 4 Weeks"

def snake_to_title(s):
    """Convert snake_case to Title Case"""
    return " ".join(word.capitalize() for word in s.replace(".md", "").split("_"))

def generate_py_file(md_path, py_path, folder_name):
    course_title = snake_to_title(os.path.basename(md_path))
    module_title = snake_to_title(folder_name) + " Cheatsheet"
    category = snake_to_title(folder_name)
    
    url_file_name = os.path.basename(md_path).replace(".md", "")
    url = f"/learn/courses/{folder_name}/{url_file_name}"
    
    template = f'''import os

def get_markdown_content():
    """Reads and returns the latest content of the markdown file"""
    md_path = os.path.join(os.path.dirname(__file__), "{os.path.basename(md_path)}")
    with open(md_path, "r", encoding="utf-8") as f:
        return f.read()

course_data = {{
    "course_title": "{course_title}",
    "module_title": "{module_title}",
    "category": "{category}",
    "url": "{url}",
    "imageUrl": "{IMAGE_URL}",
    "date": "{DURATION}",
    "body": get_markdown_content()
}}
'''
    with open(py_path, "w", encoding="utf-8") as f:
        f.write(template)
    print(f"Generated: {py_path}")


for root, dirs, files in os.walk(BASE_DIR):
    # Filter only .md files
    md_files = [f for f in files if f.endswith(".md")]
    for idx, md_file in enumerate(md_files):
        md_path = os.path.join(root, md_file)
        py_path = os.path.join(root, f"{idx}.py")  # numeric naming
        folder_name = os.path.basename(root)
        generate_py_file(md_path, py_path, folder_name)
```

```python
import os
import importlib.util

BASE_DIR = "cheatsheets"

def load_py_course_data(py_path):
    """Dynamically load a .py file and return its course_data dictionary"""
    spec = importlib.util.spec_from_file_location("course_module", py_path)
    module = importlib.util.module_from_spec(spec)
    spec.loader.exec_module(module)
    return module.course_data

def get_all_courses(base_dir=BASE_DIR):
    """Read all .py files in cheatsheets and return a list of dictionaries"""
    courses_list = []

    for root, dirs, files in os.walk(base_dir):
        py_files = [f for f in files if f.endswith(".py")]
        for py_file in py_files:
            py_path = os.path.join(root, py_file)
            try:
                course_data = load_py_course_data(py_path)
                course_dict = {
                    "category": course_data["category"],
                    "title": course_data["module_title"],
                    "date": course_data["date"],
                    "imageUrl": course_data["imageUrl"],
                    "url": course_data["url"].replace("/learn/courses/", "/cheatsheets/")
                }
                courses_list.append(course_dict)
            except Exception as e:
                print(f"Error loading {py_path}: {e}")

    return courses_list

courses = get_all_courses()
for c in courses:
    print(c)
```

```python
import os
import importlib.util
import random

BASE_DIR = "cheatsheets"

def load_py_course_data(py_path):
    """Dynamically load a .py file and return its course_data dictionary"""
    spec = importlib.util.spec_from_file_location("course_module", py_path)
    module = importlib.util.module_from_spec(spec)
    spec.loader.exec_module(module)
    return module.course_data

def get_article_with_related(category, blog):
    """
    Returns the course data for a specific article and a list of related posts from the same folder.
    
    Args:
        category (str): folder name
        blog (str): markdown file name without extension, e.g., 'feature_scaling'
    
    Returns:
        dict: result dictionary containing 'article' and 'relatedPosts'
    """
    folder_path = os.path.join(BASE_DIR, category)
    if not os.path.exists(folder_path):
        print(f"Category folder '{category}' does not exist.")
        return None

    constructed_url = f"/learn/courses/{category}/{blog}"

    article_data = None
    all_courses = []

    # Load all .py files in the folder
    py_files = [f for f in os.listdir(folder_path) if f.endswith(".py")]
    for py_file in py_files:
        py_path = os.path.join(folder_path, py_file)
        try:
            course_data = load_py_course_data(py_path)
            all_courses.append(course_data)
            if course_data.get("url") == constructed_url:
                article_data = course_data
        except Exception as e:
            print(f"Error loading {py_path}: {e}")

    if not article_data:
        print(f"No article found for URL: {constructed_url}")
        return None

    # Prepare related posts (randomly pick 3 different from main article)
    other_courses = [c for c in all_courses if c.get("url") != constructed_url]
    related_posts = []
    sample_count = min(3, len(other_courses))
    if sample_count > 0:
        for c in random.sample(other_courses, sample_count):
            related_posts.append({
                "title": c.get("module_title", ""),
                "category": c.get("category", ""),
                "date": c.get("date", ""),
                "url": c.get("url", "").replace("/learn/courses/", "/cheatsheets/")
            })

    # Build result dictionary
    result = {
        "article": {
            "title": article_data.get("module_title", ""),
            "category": article_data.get("category", ""),
            "date": article_data.get("date", ""),
            "readTime": article_data.get("readTime", ""),  # optional field
            "imageUrl": article_data.get("imageUrl", ""),
            "body": article_data.get("body", "")
        },
        "relatedPosts": related_posts
    }

    return result

# Example usage
res = get_article_with_related("interview_questions", "computer_vision")

```
